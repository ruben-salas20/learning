```python
# Start coding here
# Use as many cells as you need

def validate_user(name, email, password):

    if not validate_password(password):
        raise ValueError("Contraseña ingresado no cumple.")

    if not validate_email(email):
        raise ValueError("Correo ingresado no cumple.")

    if not validate_name(name):
        raise ValueError("Nombre ingresado no cumple.")

    return True

    

def register_user(name, email, password):

    try:
        if validate_user(name, email, password) == True:
            return {"name": name, "email": email, "password": password}
    except ValueError:
        return False

print(register_user("Ruben", "ruben@gmail.com", "Salas123"))
```
